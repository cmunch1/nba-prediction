## Feature Selection

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 500)

from sklearn.feature_selection import SelectKBest, chi2


import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path  #for Windows/Linux compatibility
DATAPATH = Path(r'data')
```

```python
train = pd.read_csv(DATAPATH / "train_features.csv")
test = pd.read_csv(DATAPATH / "test_features.csv")
```

```python
#remove 2nd to last season from test set
#it was needed to just generate rolling mean that began in previous season

latest_season = test['SEASON'].unique().max()

test = test[test['SEASON'] >= (latest_season)]

```

```python
train.to_csv(DATAPATH / "train_selected.csv",index=False)
test.to_csv(DATAPATH / "test_selected.csv",index=False)
```
